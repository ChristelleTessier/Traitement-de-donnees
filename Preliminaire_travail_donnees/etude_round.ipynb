```python
import pandas as pd
```

## Création des fonctions

```python
def chercher_info_tournoi_general(data, round):

    # Récupérer la liste des tournois dont round correspond
    data_test = data[data['round'] == round]


    # Création d'un DataFrame pour stocker les informations
    info = pd.DataFrame(columns=['tournoi_id', 'nb_match', 'type_round'])

    # Boucle sur chaque tournoi unique
    for tournoi_id in data_test['tourney_id'].unique():
        matchs_tournoi = data[data['tourney_id'] == tournoi_id]
        nb_match = matchs_tournoi.shape[0]
        type_round = matchs_tournoi['round'].unique()

        # Ajoute une ligne au DataFrame
        info = pd.concat([info, pd.DataFrame([{
            'tournoi_id': tournoi_id,
            'nb_match': nb_match,
            'type_round': type_round
        }])], ignore_index=True)

    return info
```

```python
def chercher_type_round(info, nb_match):
    unique_rounds = set()
    info = info[info['nb_match'] == nb_match]
    for round in info['type_round']:
        unique_rounds.update(round)

    return unique_rounds
```

```python
def liste_round(info):
    # Création d'un DataFrame pour stocker les informations
    rep = pd.DataFrame(columns=['nb_match', 'type_round'])

    for nb_match in sorted(info['nb_match'].unique()):

        type_round = chercher_type_round(info, nb_match)

        # Ajoute une ligne au DataFrame
        rep = pd.concat([rep, pd.DataFrame([{
            'nb_match': nb_match,
            'type_round': type_round
        }])], ignore_index=True)


    return rep
```

```python
def chercher_info_tournoi(genre,tournoi_id, type):

    fichier = "../donnees_tennis/"
    fichier += genre.upper() + "/"
    fichier += genre + "_" + "matches_"
    if type !="":
        fichier += type + "_"
    fichier += tournoi_id[0:4] + ".csv"
    data = pd.read_csv(fichier)
    data = data[data['tourney_id'] == tournoi_id]
    data_resul = data[['tourney_id',"match_num","round","winner_name","loser_name"]]

    return data_resul
```
