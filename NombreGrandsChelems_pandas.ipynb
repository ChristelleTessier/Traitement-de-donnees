```python
# Module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
chemin = os.path.join("Donnees","ATP_match_simple.csv")
matchs = pd.read_csv(chemin)
matchs.head()
```

```python
matchs.shape
```

```python
grand_chelem = matchs[matchs['tourney_level']=='G']
grand_chelem = grand_chelem[grand_chelem['round']=='F']
grand_chelem.head()
```

```python
grand_chelem.shape
```

```python
plus_10_victoires = grand_chelem['winner_name'].value_counts().reset_index()
plus_10_victoires = plus_10_victoires[plus_10_victoires['winner_name']>9]
plus_10_victoires = plus_10_victoires.rename(columns={'index': 'Joueur', 'winner_name': 'Nombre'})
plus_10_victoires
```

```python
plt.bar(plus_10_victoires['Joueur'], plus_10_victoires['Nombre'])
```
